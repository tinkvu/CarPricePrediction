```python
import pandas as pd

#df = pd.read_csv('/content/TravelRecommenderUsingLLM/Raw Data/detailed_reviews_Dublin.csv')
df=pd.read_csv('/content/detailed_reviews_Dublin.csv')
df.head()
```

```python
df.info()
```

```python
unique_place_count = df['place_name'].unique().size
print(f"Number of unique place names: {unique_place_count}")
```

```python
# prompt: get not null datas of this column : df['review_translated_text']

df_not_null = df[df['review_translated_text'].notnull()]

# Get the number of rows with not null values in 'review_translated_text'
number_of_not_null_rows = df_not_null.shape[0]

# Print the not null data
print(df_not_null['review_translated_text'])
```

```python
cols=['place_name','rating','review_text','review_translated_text']
df=df[cols]
df
```

```python
df['review'] = df['review_translated_text'].combine_first(df['review_text'])
df=df[['place_name','rating','review']]
df
```

```python
# Combine reviews with the same place_name
df_combined = df.groupby('place_name').agg({
    'rating': 'mean',  # Calculate average rating
    'review': lambda x: ' '.join(str(review) for review in x)  # Join all reviews into a single string
}).reset_index()
df_combined.head()
```

```python
df= df_combined
```

```python
df.to_csv('processed_datas.csv')
```
